# 4. Int Değişkenler / Klavyeden Veri Girişi

* C#'da bir değişkeni int'e dönüştürmenin -sayının boyutuna göre- üç farklı komutu  vardır; 
    * <code>ToInt16</code>
    * <code>ToInt32</code>, en yaygın kullanılanıdır. 
    * <code>ToInt64</code>

### Örnek: Dönüşüm

<img src = "https://i.hizliresim.com/feq9viu.png">

* TextBox'dan gelen değer stringdir ve integera dönüştürülmesi gerekir.
    * <code>Convert.ToInt16(textBox1.Text)</code> ile integera dönüştürülüp int değişkene atama yapılır.
* Label string değeri gösterebildiği için int değişken tekrar stringe dönüştürülmesi gerekir.
    * <code>sayi.ToString()</code> ile dönüştürülür.
