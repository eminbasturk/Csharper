# C# Booelans

Çoğu zaman, programlamada, aşağıdaki gibi iki değerden yalnızca birine sahip olabilen bir veri türüne ihtiyacınız olacaktır: 
* YES / NO
* ON / OFF
* TRUE / FALSE

Bunun için C#, <code>true</code> veya <code>false</code> değerlerini alabilen bir <code>bool</code> veri türüne sahiptir.
