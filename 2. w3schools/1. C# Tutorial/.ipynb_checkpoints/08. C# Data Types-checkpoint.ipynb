# Data Types

Değişkenler bölümünde açıklandığı gibi, C#'ta bir değişken belirtilen bir veri türü olmalıdır: 



Bir veri türü, değişkenin değerlerinin boyutunu ve türünü belirtir. 

Karşılık gelen değişken için doğru veri türünün kullanılması önemlidir; hataları önlemek, zamandan ve bellekten tasarruf etmek için, ama aynı zamanda kodunu daha bakımlı ve okunabilir hale de getirir. En yaygın veri türleri şunlardır: 

| **Data Type** | **Size** | **Description** |
| --- | --- | --- |
| int | 4 bytes | -2.147.483.648 ile 2.147.483.647 arasındaki tamsayıları depolar |
| long | 8 bytes | -9.223.372.036.854.775.808 ile 9.223.372.036.854.775.807 arasındaki tamsayıları depolar |
| float | 4 bytes | Kesirli sayıları depolar. 6 ila 7 ondalık basamak depolamak için yeterlidir |
| double | 8 bytes | Kesirli sayıları depolar. 15 ondalık basamak depolamak için yeterlidir |
| bool | 1 bit | true veya false değerleri depolar | 
| char | 2 bytes | Tek tırnak içine alınmış tek bir karakter/harf depolar | 
| string | karakter başına 2 bytes | Çift tırnak içine alınmış bir karakter dizisini depolar | 



# Numbers

Sayı türleri iki gruba ayrılır: 

**Integer types**, pozitif veya negatif (123 veya -456 gibi) tam sayıları ondalık olmadan depolar. Geçerli türler <code>int</code> ve <code>long</code>'dur. Hangi türü kullanman gerektiği, sayısal değere bağlıdır.

**Floating point types**, bir veya daha fazla ondalık basamak içeren kesirli kısmı olan sayıları temsil eder. Geçerli türler <code>float</code> ve <code>double</code>'dır. 

C#'ta bir çok sayısal tür olmasına rağmen, sayılar için en çok kullanılanlar <code>int</code> (tamsayılar için) ve <code>double</code>'dır (ondalıklı sayılar için). Ancak, sen okumaya devam ettikçe hepsi açıklanacak.

## Integer Types

### Int
<code>int</code> veri tipi, -2147483648 ile 2147483647 arasındaki tam sayıları saklayabilir. Genel olarak tutorialda sayısal bir değere sahip değişkenler oluşturduğumuzda <code>int</code> veri tipi tercih edilen veri tipidir.

## Floating Point Types

9.99 veya 3.14515 gibi ondalık basamaklı bir sayıya ihtiyacın olduğunda ondalıklı sayı türünü kullanmalısın.

<code>float</code> veya <code>double</code> veri türleri kesirli sayıları saklayabilir. Değeri floatlar için "F" ve doublelar için "D" ile bitirmen gerektiğini unutma: 



#### <code>float</code> mı <code>double</code> mı kullanılmalı?

Ondalıklı bir sayının kesinliği (**precision**) noktadan sonra kaç basamak olabileceğini gösterir. <code>float</code>'ın kesinliği yalnızca altı veya yedi ondalık basamakken, <code>double</code> değişkenlerin kesinliği yaklaşık 15 basamaktır. Bu nedenle, çoğu hesaplama için <code>double</code> kullanmak daha güvenlidir.
