# Variables

Değişkenler, veri değerlerini depolamak için konteynerlardır. 

C#'ta farklı türde (**types**) değişkenler vardır (farklı anahtar sözcüklerle tanımlanır), örneğin: 
* <code>int</code>, 123 veya -123 gibi tam sayıları (numaraları) ondalık olmadan saklar
* <code>double</code>, 19.99 veya -19.99 gibi ondalık basamaklı ondalıklı sayıları saklar
* <code>char</code>, 'a' veya 'B' gibi tek karakterleri saklar. Karakter değerleri tek tırnak içine alınır
* <code>string</code>, "Hello World" gibi metinleri depolar. String değerleri çift tırnak içine alınır
* <code>bool</code>, değerleri iki durumla saklar: doğru veya yanlış (true or false)

# Constants
Sabitler

Başkalarının (veya kendinin) mevcut değerlerin üzerine yazmasını istemiyorsan, <code>const</code> keywordünü değişken türünün önüne ekleyebilirsin.

Bu, değişkeni değiştirilemez ve salt okunur anlamına gelen "constant" (sabit) olarak ilan edecektir:



<code>const</code> keywordü, bir değişkenin her zaman aynı değeri depolamasını istediğinde kullanışlıdır, böylece başkaları (veya kendin) kodunu bozmaz. Genellikle sabit olarak adlandırılan bir örnek, PI'dir (3.14159...).

**Note**: Değer atamadan bir sabit değişken tanımlayamazsın. Bunu yaparsan bir hata oluşur: *Bir const alanı bir değerin sağlanmasını gerektirir.*

# Identifiers

Tüm C# değişkenleri (**variables**) benzersiz adlarla (**unique names**) tanımlanmalıdır (**identified**).

Bu benzersiz adlara tanımlayıcı (**identifiers**) denir.

Identifierlar kısa adlar (x ve y gibi) veya daha açıklayıcı adlar (age, sum, totalVolume) olabilir.

**Note**: Anlaşılabilir ve sürdürürebilir kod oluşturmak için açıklayıcı adların kullanılması önerilir: 



Değişkenleri adlandırmak için genel kurallar şunlardır: 

* İsimler harf, rakam ve alt çizgi karakteri (_) içerebilir.
* İsimler bir harf ile başlamalıdır.
* İsimler küçük harfle başlamalı ve boşluk içermemelidir.
* İsimler büyük/küçük harfe duyarlıdır. ("myVar" ve "myvar" farklı değişkenlerdir)
* Zaten C#'da kullanılan kelimeler (<code>int</code>, <code>double</code> gibi) isim olarak kullanılamaz.
