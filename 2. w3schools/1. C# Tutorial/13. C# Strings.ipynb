# Strings - Special Characters

Stringlerin tırnak içinde yazılması gerektiğinden, C# aşağıdaki stringi yanlış anlayacak ve bir hata üretecektir: 



Bu sorunu önlemenin çözümü, ters eğik çizgi kaçış karakterini kullanmaktır (**backslash escape character**).

Ters eğik çizgi (<code>\</code>) kaçış karakteri, özel karakterleri string karakterlerine dönüştürür: 

| **Escape character** | **Result** | **Description** | 
| --- | --- | --- |
| \\' | ' | Tek tırnak |
| \\" | " | Çift tırnak |
| \\\ | \ | Ters eğik çizgi |

<code>\\"</code> dizisi, bir stringe çift tırnak ekler: 



<code>\\'</code> dizisi, bir stringe tek tırnak ekler: 



<code> \\\ </code> dizisi, bir stringe ters eğik çizgi ekler: 



C#'taki diğer yararlı kaçış karakterleri (escape characters) şunlardır: 

| **Code** | **Result** |
| --- | --- |
| \\n | Yeni Satır |
| \\t | Tab | 
| \\b | Boşluk |
