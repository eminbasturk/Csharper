# 1. Merhaba Dünya

* Uygulama geliştirerek ilerleneceği için öncelikle bazı temel araçların bir form üzerinden öğrenilmesi gerekiyor.
* Bu yüzden projelerimizi **Windows Form Application** olarak açıyoruz.
* Temel araçlar, programımızın ekranında bulunan, görsel ve tasarımsal olarak yarar sağlayan hem de kodumuzun işlevlerini gerçekleştiren araçlardır.
* **Toolbox**: Araçların bulunduğu sekmedir.
* **Properties**: Formun özelliklerini gösterir. Formun özelliklerini bu sekmeden değiştirebiliriz.
* Bir form kenar ve köşelerinden çekilerek büyütülebilir.
* Formun başlığını değiştirmek istiyorsan *Properties*'den **Text** kısmını değiştirebilirsin.
* Formun arka plan rengini değiştirmek için **BackColor** özelliğini kullanabilirsin.
* Formun içerisine *Toolbox*'dan araç sürükleyebilirsin.
* **Solution Explorer**, projenin görüntülendiği sekmedir.

## <code>Button</code>
* Olayları tetiklemek için kullanılan araçtır. Üzerine tıklandığı zaman çalışır.
* **Button**'ın ismini de *Text* özelliğinden değiştirebilirsin. Aynı zamanda **Font** özelliğinden yazı tipini, stilini ve boyutunu da değiştirebilirsin.
* Eğer Button'a çift tıklarsan onu tetiklemiş olursun ve kod sayfasına gidersin.
    * C#'da bir kütüphaneyi **using** komutu ile kullanabilirsin.
        * Başında **System.** bulunan kütüphaneler sistem kütüphaneleridir ve otomatik olarak eklenir.
    * **namespace**, projenin ismini tutar.
    * *Button*'a tıkladıktan sonra yapılmasını istediğin şeyi **Click** metoduna yazarsın.
    * Programlamada sıra komutlar birbirlerine "**.**" ile bağlanır.

<img src ="https://i.hizliresim.com/taljcul.png">

## <code>MessageBox</code>
* Mesaj kutusunda bir metin göstermek istersen;
* <code>MessageBox.Show("Merhaba Dünya");</code>
    * Eğer bir komutun solunda küp semboli varsa o komut metottur ve parantez ile yazılır.
    * Metinler çift tırnak arasına yazılır.
    * Komut satırı "**;**" ile sonlanır.

<img src = "https://i.hizliresim.com/3epvi9i.png">

# 2. Label ve TextBox Kullanımı

## <code>Label</code>

* Bilgi vermek amacıyla kullanılan etiketlerdir.
    * İnternet üzerinde bir siteye kayıt olurken kullanıcı adını ve şifresini nereye yazacağını göstermek amacıyla kullanılır.

### Not

* Otomatik tamamlama kapanırsa **Ctrl + Space** kullan.
* Aracın özelliğini Properties kısmından Text kullanarak değiştirdiğin gibi kodu yazarken de yine **Text** kullanarak değiştirirsin.

<img src = "https://i.hizliresim.com/4f1bk9q.png">

<img src = "https://i.hizliresim.com/hxqcgpr.png">


## <code>TextBox</code>

* Klavyeden veri alır.

### Örnek
Bir TextBox, bir Button ve bir Labelin olsun. Button'a tıklandığında TextBox'a yazdığın bilgi Label'a yazdırılsın.

<img src = "https://i.hizliresim.com/e0t6s7z.png">

* Nereye yazdırılacaksa orası sol tarafta olur.
* Nereden ya da ne yazacaksan orası sağ tarafta olur.

<img src = "https://i.hizliresim.com/1680kks.png">

### Not

* **C#'da yorum satırı // ile açılır.**

### Alıştırma
3 tane TextBox, 3 tane Label olsun. Buttona basıldığında TextBox'daki bilgiler labellara yazdırılsın.

# 3. ComboBox ve ListBox Araçları Kullanımı

## <code>ComboBox</code>
* Değerleri açılan bir menü içerisinden seçmeye olanak tanıyan bir araçtır.
* Değerleri aracın üstündeki ikona tıklayıp ardından **Edit Items** diyerek açılan *String Collection Editor* penceresinden ekleyebilirsin.
* Değerler satır satır girilir.
* ComboBox'a kod kısmından da eleman ekleyebilirsin.

### Örnek
Buttona tıklandığında ComboBox'a şehir eklensin.

<img src = "https://i.hizliresim.com/oww4eov.png">

<img src = "https://i.hizliresim.com/5j2a0gn.png">

### Not

* ComboBox'a eleman eklerken doğrudan Text kullanılmaz çünkü elemanın **ekle**nmesi gerekiyor.
* Formun **Font** değerini değiştirdiğin zaman sonradan eklediğin her tool için defaultu o olur.

### Alıştırma
Buttona tıklandığında TextBox'a girilen metni ComboBox'a eklesin.

<img src = "https://i.hizliresim.com/i96uhgk.png">

<img src = "https://i.hizliresim.com/hncnjoi.png">

## <code>ListBox</code>
* ComboBox'ın açılmış hali gibidir. 
* Elemanlar ComboBox'a eklendiği gibi eklenir.

### Alıştırma

Buttona tıklandığında kişinin adını, soyadını, yaşını, mesleğini ve maaş bilgilerini TextBox'dan alıp, bunları tek sıra halinde bir ListBox'a her bir kişi bir satıra gelecek şekilde bir program yaz.

* Forma ekledikten sonra **Set Musk...** tıklarsan *Input Mask* ekranı açılır, seçimini buradan yapabilirsin.

<img src = "https://i.hizliresim.com/458m6kg.png">

* *Size Mode* kısmından resmi sığdırma, doldurma gibi işlemleri gerçekleştirebilirsin. 
* **StretchImage**

# 5. CheckBox, RadioButton, RichTextBox, Panel ve GroupBox Kullanımı

## <code>CheckBox</code>
* Seçilen değerin alınıp alınmadığını kontrol eder. 
* True veya False durumları vardır. 
* Kullanıcıya çoklu seçim yaptırmada kullanılır. 
    * Hamburger malzemelerini seçmek gibi.
    
## <code>GroupBox</code>
* Araçları bir arada tutmak için kullanılır.
    * Malzemeler isminde bir GroupBox açıp içerisine marul, domates isimlerinde CheckBoxlar ekleyebiliriz.
    
<img src = "https://i.hizliresim.com/odm5df9.png">    

## <code>RadioButton</code>

* CheckBoxlar gibi seçim için eklenir. Ancak yalnızca bir tanesine tıklanabilir. 
* Genelde ikili seçimlerde kullanılır.
    * Cinsiyet gibi.
    
<img src = "https://i.hizliresim.com/37mv1c8.png">

## <code>Panel</code>

* Araçları bir araya getirmeye yarar. 
* Daha çok tasarımsal olarak kullanılır.
* GroupBox ile arasındaki fark çerçevesinin gözükmemesidir.

<img src = "https://i.hizliresim.com/3lece3d.png">

## <code>RichTextBox</code>

* TextBox'ın zengin, genişletilmiş ve büyütülmüş halidir.

<img src = "https://i.hizliresim.com/hv1bshm.png">

İki modülü olacak. Sol tarafında uçuş bilgileri, sağ tarafında yolcu bilgileri olacak. Uçuş bilgilerinde kaydet butonuna basıldığında bilgilerin tamamını alt tarafta bulunan listboxa yazacak.

<img src = "https://i.hizliresim.com/rzk803o.png">

<img src = "https://i.hizliresim.com/s815q2o.png">

### Not:

* ListBox'ın yatayda çok fazla uzadığında alan dışına çıkanların görünmesi için *Properties* kısmından **HorizontalScrollBar** özelliğini *True* yap.
* Eğer bir aracın uygulama çalıştırıldığında görünmemesini istiyorsan **Visible** özelliğini *False* yap.
* Eğer iki aracın textini birbirleriyle değiştirmek istiyorsan bir ara araç veya değişken oluştur, bir aracın textini diğerine atarken diğerinin texti geçiş aracında olsun.
