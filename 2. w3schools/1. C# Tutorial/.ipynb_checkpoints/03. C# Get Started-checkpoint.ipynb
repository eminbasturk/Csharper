# C# Get Started
*C# Başlarken*

## C# IDE

* C# ile başlamanın en kolay yolu bir IDE kullanmaktır.
* Kodu düzenlemek ve derlemek için bir IDE (Entegre Geliştirme Ortamı) kullanılır.
* Eğitimde Visual Studio Community kullanılıyor.
* C# ile yazılmış uygulamalar .NET Framework kullanır, bu nedenle program, framework ve dil Microsoft tarafından oluşturulduğundan Visual Studio'yu kullanmak mantıklıdır.

## C# Install
*C# Kurulumu*

* Visual Studio Installer indirip yükledikten sonra .NET workloadını seç ve **Modify/Install** butonuna tıkla: 

<img src = "https://www.w3schools.com/cs/Workload.png">

* Kurulum tamamlandıktan sonra, Visual Studio'yu kullanmaya başlamak için **Launch** butonuna tıkla.
* Başlangıç penceresinde **Create a new project**i seç: 

<img src = "https://www.w3schools.com/cs/Newproject.png">

* Ardından "Daha fazla araç ve özellik yükle" butonuna tıkla: 

<img src = "https://www.w3schools.com/cs/Newproject2.png">

* Listeden "Console App (.NET Core)" öğesini seç ve Next butonuna tıkla: 

<img src = "https://www.w3schools.com/cs/Consoleapp.png">

* Projen için bir isim gir ve Create butonuna tıkla:

<img src = "https://www.w3schools.com/cs/saveas.png">

* Visual Studio, projen için otomatik olarak bazı kodlar üretecektir: 

<img src = "https://www.w3schools.com/cs/Program.png">

* Kod şöyle bir şeye benzemelidir: 



* Yukarıdaki kodu anlamadıysan endişelenme - bu sonraki bölümlerde ayrıntılı olarak tartışılacak. Şimdilik, kodu nasıl çalıştıracağına odaklan.

* Klavyendeki **F5** düğmesine basarak programı çalıştır (veya "**Debug**" -> "**Start Debugging**" seçeneğine tıkla). Bu, kodunu derleyecek ve yürütecek, sonuç şuna benzer: 



* **Tebrikler**! Artık ilk C# programını yazın ve çalıştırdın.
