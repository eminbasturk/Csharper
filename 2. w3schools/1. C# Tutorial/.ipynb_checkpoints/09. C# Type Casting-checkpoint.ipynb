# Type Casting

Tip dönüşümü, bir veri tipinin değerini başka bir tipe atanmasıdır.

C#'ta dönüşümün iki türü vardır:

* **Implicit Casting** (otomatik olarak) - daha küçük boyutlu bir veri türünü daha büyük bir boyutlu veri türüne dönüştürme
    * <code>char</code> -> <code>int</code> -> <code>long</code> -> <code>float</code> -> <code>double</code> 
    
* **Explicit Casting** (manuel olarak) - daha büyük boyutlu bir veri türünü daha küçük bir boyutlu veri türüne dönüştürme
    *  <code>double</code> -> <code>float</code> -> <code>long</code> -> <code>int</code> -> <code>char</code>
