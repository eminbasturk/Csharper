# Conditions and If Statements

C# matematiğin olağan mantıksal koşullarını destekler:

* Küçüktür: <code>a < b</code>
* Küçük eşittir: <code>a <= b</code> 
* Büyüktür: <code>a > b</code>
* Büyük eşittir: <code>a >= b</code>
* Eşittir: <code>a == b</code>
* Eşit değildir: <code>a != b</code>

Farklı kararlar için farklı eylemler gerçekleştirmek için bu koşulları kullanabilirsin.
    
C# aşağıdaki koşullu ifadelere sahiptir:
    
* Belirtilen koşul doğruysa yürütülecek bir kod bloğu belirtmek için <code>if</code> kullan.
* Aynı koşul yanlışsa yürütülecek bir kod bloğu belirtmek için <code>else</code> kullan.
* İlk koşul yanlışsa test edilecek yeni bir koşul belirtmek için <code>else if</code> kullan.
* Yürütülecek birçok alternatif kod bloğu belirtmek için <code>switch</code> kullan.

# The if Statement

Bir koşul <code>True</code> ise yürütülecek bir C# kodu bloğu belirtmek için <code>if</code> ifadesini kullan.



<code>if</code>'in küçük harflerle yazıldığına dikkat et. Büyük harfler (If veya IF) bir hata oluşturur.

Aşağıdaki örnekte, 20'nin 18'den büyük olup olmadığını anlamak için iki değeri test ediyoruz. Koşul <code>True</code> ise, biraz text yazdır: 



Değişkenleri de test edebiliriz: 



Yukarıdaki örnekte, **x**'in **y**'den büyük olup olmadığını test etmek için (<code>></code> operatörünü kullanarak) iki değişken, **x** ve **y**yi kullanıyoruz. x = 20 ve y = 18 olduğuna göre ve 20'nin 18'den büyük olduğunu bildiğimiz için ekrana "x is greater than y" yazıyoruz.

# The else Statement

* Koşul <code>False</code> ise yürütülecek bir kod bloğu belirtmek için <code>else</code> ifadesini kullan.





Yukaridaki örnekte time (20), 18'den büyüktür, dolayısıyla koşul <code>False</code>'dir. Bu nedenle <code>else</code> koşuluna geçilir ve ekrana "Good evening" yazdırılır. Time 18'den az ise, program "Good day" yazdırır.

# The else if Statement

İlk koşul <code>False</code> ise yeni bir koşul belirtmek için <code>else if</code> ifadesini kullan.





Yukarıdaki örnekte, time (22) 10'dan büyüktür, dolayısıyla **ilk koşul** <code>False</code> olur. <code>else if</code> ifadesindeki bir sonraki koşul da <code>False</code>'dir, bu nedenle hem **condition1** hem de **condition2**'nin ikisi de <code>False</code> olduğu için <code>else</code> koşulna geçeriz ve ekrana "Good evening" yazdırırız.

Ancak saat 14 olsaydı programımız "Good day." yazardı.
