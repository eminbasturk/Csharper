# For Loop

Bir kod bloğunda tam olarak kaç kez döngü yapmak istediğinizi biliyorsan, <code>while</code> döngüsü yerine <code>for</code> döngüsünü kullan:



**Statement 1**, kod bloğunun yürütülmesinde önce yürütülür (bir kez).

**Statement 2**, kod bloğunu yürütme koşulunu tanımlar.

**Statement 3**, kod bloğu yürütüldükten sonra (her seferinde) yürütülür.

Aşağıdaki örnek, 0 ile 4 arasındaki sayıları yazdıracaktır:



Statement 1, döngü başlamadan önce bir değişken ayarlar (<code>int i = 0</code>).

Statement 2, döngünün çalışması için koşulu tanımlar (<code>i</code>, <code>5</code>'den küçük olmalıdır). Koşul <code>true</code> ise döngü baştan başlar, <code>false</code> ise döngü sona erer.

Statement 3, döngüdeki kod bloğu her yürütüldüğünde bir değeri (<code>i++</code>) artırır.
