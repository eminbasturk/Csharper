## Arithmetic Operators

Aritmetik operatorler yaygın matematiksel işlemleri gerçekleştirmek için kullanılır:

| **Operator** | **Name** | **Description** | **Example** |
| --- | --- | --- | --- |
| + | Addition (Toplama) | İki değeri bir araya toplar | x + y |
| - | Subtraction (Çıkarma) | Bir değeri diğerinden çıkarır | x - y |
| * | Multiplication (Çarpma) | İki değeri çarpar | x * y |
| / | Division (Bölme) | Bir değeri diğerine böler | x / y |
| % | Modulus (Mod Alma) | Bölme işleminin kalanını verir | x % y |
| ++ | Increment (Artma) | Bir değişkenin değerini 1 artırır | x++ |
| -- | Decrement (Azalma) | Bir değişkenin değerini 1 azaltır | x-- |

## Assignment Operators

Atama operatörleri, değişkenlere değer atamak için kullanılır.

Aşağıdaki örnekte, **x** adlı bir değişkene **10** değerini atamak için atama (**assignment**) operatörünü (<code>=</code>) kullanıyoruz: 



Toplama atama (**addition assignment**) operatörü (<code>+=</code>) bir değişkene bir değer ekler: 



Tüm atama operatörlerinin listesi: 

| **Operator** | **Example** | **Same As** | **Note**
| --- | --- | --- | --- |
| = | x = 5 | x = 5 | --- |
| += | x += 3 | x = x + 3 | --- |
| -= | x -= 3 | x = x - 3 | --- |
| *= | x *= 3 | x = x * 3 | --- |
| %= | x %= 3 | x = x % 3 | --- |
| &= | x &= 3 | x = x & 3 | Bit düzeyinde VE operatörü ile karşılaştırma yapar. Sonuç tam sayıdır. |
| ^= | x ^= 3 | x = x ^ 3 | Bit düzeyinde XOR operatörü ile karşılaştırma yapar. Sonuç tam sayıdır. |
| >>= | x >>= 3 | x = x >> 3 | Bit düzeyinde sağa kaydırma yapılır. Kalan değerler 0 ile dolar. |
| <<= | x <<= 3 | x = x << 3 | Her bir sola kaydırma sayıyı 2 ile çarpmak gibidir. | 

## Comparison Operators

Karşılaştırma operatörleri, iki değeri (veya değişkeni) karşılaştırmak için kullanılır. Bu, programlamada önemlidir, çünkü cevapları bulmamıza ve karar vermemize yardımcı olur.

Bir karşılaştırmanın dönüş değeri <code>True</code> veya <code>False</code>'dir. Bu değerler *Boolean değerleri* olarak bilinir.

Aşağıdaki örnekte, 5'in 3'ten büyük olup olmadığını bulmak için büyüktür operatörünü (**greater than operator**) (<code>></code>) kullanıyoruz: 



Tüm karşılaştırma operatörlerinin listesi: 

| **Operator** | **Name** | **Example** |
| --- | --- | --- |
| == | Eşittir | x == y |
| != | Eşit değil | x != y | 
| > | Büyüktür | x > y |
| < | Küçüktür | x < y |
| >= | Büyük veya eşittir | x >= y |
| <= | Küçük veya eşittir | x <= y |
