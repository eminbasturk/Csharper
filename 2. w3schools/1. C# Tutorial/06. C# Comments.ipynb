# C# Comments

Comments (Yorumlar), C# kodunu açıklamak ve daha okunaklı hale getirmek için kullanılabilir. Alternatif kodu test ederken yürütmeyi önlemek için de kullanılabilir.

## Single-line Comments
Tek Satırlık Yorumlar

Tek satırlık yorumlar iki eğik çizgi (<code>//</code>) ile başlar.

<code>//</code> ile satırın sonu arasındaki herhangi bir metin C# tarafından yoksayılır (yürütülmez).

Çok Satırlı Yorumlar

Çok satırlı yorumlar <code>/*</code> ile başlar ve <code> */ </code> ile biter.

<code>/*</code> ve <code> */ </code> arasındaki tüm metinler C# tarafından yok sayılır.

Hangisini kullanmak istediğin sana kalmış. Normalde, kısa yorumlar için <code>//</code> ve daha uzun yorumlar için <code>/* */</code> kullanılır.
