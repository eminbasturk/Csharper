# Parameters and Arguments

Bilgi methodlara parametre olarak iletilebilir. Parametreler, method içinde değişkenler olarak işlev görür.

Method adından sonra parantez içinde belirtilirler. İstediğin kadar parametre ekleyebilirsin, sadece virgülle ayır.

Aşağıdaki örnek, parametre olarak **fname** adlı bir <code>string</code> alan bir methoda sahiptir. Method çağrıldığında, methodun içinde tam adı yazdırmak içi kullanılan bir ad iletiriz: 



Methoda bir **parameter** iletildiğinde buna **argument** denir. Dolayısıyla, yukarıdaki örnekten: <code>fname</code> bir **parameter**dir, <code>Liam</code>, <code>Jenny</code> ve <code>Anja</code> ise **argument**lerdir.

## Default Parameter Value

Eşittir işaretini (<code>=</code>) kullanarak varsayılan bir parametre değeri de kullanabilirsin.

Default değere sahip bir parametre genellikle "**optional parameter**" olarak bilinir. Yukarıdaki örnekte <code>country</code> isteğe bağlı bir parametredir ve <code>"Norway"</code> default değerdir.
