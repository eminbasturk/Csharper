# Multidimensional Arrays

Şimdiye kadar öğrendiğin arrayler tek boyutlu arraylerdir. (**single dimension arrays**)

Ancak, verileri satırlar ve sütunlar içeren bir tablo gibi tablo şeklinde depolamak istiyorsan, çok boyutlu arraylere (**multidimensional arrays**) aşina olman gerekir.

Çok boyutlu bir array temel olarak bir arrayler arrayidir.

Arraylerin herhangi bir sayıda boyutu olabilir. En yaygın olanları iki boyutlu arraylerdir. (2d)

## Two-Dimensional Arrays

2D bir array oluşturmak için her arrayi kendi süslü parantezleri içine ekle ve köşeli parantezlerin içine bir virgül (<code>,</code>) ekle.



**Good to know**: Tek virgül <code>[,]</code>, arrayin two-dimensional olduğunu belirtir. Three-dimensional bir arrayde iki virgül bulunur: <code>int[,,]</code>.

<code>numbers</code> artık elemanlarının iki adet array olduğu bir arraydir. İlk array elemanı üç adet eleman içerir: 1, 4 ve 2, ikinci array elemanı ise 3, 6 ve 8'i içerir. Bunu görselleştirmek için arrayı satırları ve sütunları olan bir tablo olarak düşün: 

|  | COLUMN 0 | COLUMN 1 | COLUMN 2 |
| --- | --- | --- | --- |
| ROW 0 | 1 | 4 | 2 |
| ROW 1 | 3 | 6 | 8 |
