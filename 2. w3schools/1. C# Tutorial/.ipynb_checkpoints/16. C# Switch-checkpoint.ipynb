# Switch Statements

Yürütülecek birçok kod bloğundan birini seçmek için <code>switch</code> ifadesini kullan.



Bu nasıl çalışır: 
* <code>switch</code> ifadesi bir kez değerlendirilir
* İfadenin değeri, her <code>case</code>'in değerleriyle karşılaştırılır.
* Bir eşleşme varsa, ilgili kod bloğu yürütülür.
* <code>break</code> ve <code>default</code> keywordleri daha sonra açıklanacak.

## The break Keyword

C# bir <code>break</code> keywordüne ulaştığında, switch bloğundan kopar.

Bu, blok içinde daha fazla kod ve case testinin yürütülmesini durduracaktır.

Bir eşleşme bulunduğunda ve iş bittiğinde, break zamanı. Daha fazla teste gerek yok.

Bir break, switch bloğundaki kodun geri kalanının tamamının yürütülmesini "ignoreladığından" çok fazla yürütme süresi kazandırabilir.
