# C# Methods

**Method**, yalnızca çağrıldığında çalışan bir kod bloğudur.

Parametre olarak bilinen verileri bir methoda iletebilirsiniz.

Methodlar, belirli eylemleri gerçekleştirmek için kullanılır ve bunlar aynı zamanda **functions** olarak da bilinir.

Methodlar neden kullanılır? Kodu yeniden kullanmak için: kodu bir kez tanımla ve birçok kez kullan.

## Create a Method

Bir method, methodun adı ve ardından parantez **()** ile tanımlanır. C#, <code>Main()</code> gibi zaten aşina olduğun önceden tanımlanmış bazı methodlar sağlar, ancak belirli eylemleri gerçekleştirmek için kendi methodlarını da oluşturabilirsin: 

Program classı içinde bir method oluştur: 



* <code>MyMethod()</code> methodun adıdır.
* <code>static</code>, methodun Program classına ait olduğu ve Program classının bir nesnesi olmadığı anlamına gelir. 
* <code>void</code>, bu methodun bir dönüş değeri olmadığı anlamına gelir.

**Note**: C#'ta, kodun okunmasını kolaylaştırdığından, methodları adlandırırken büyük harfle başlamak iyi bir uygulamadır.
